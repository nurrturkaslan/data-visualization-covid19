```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("novel_corona_virus_19/covid_19_data.csv")
```

```python
df.plot()
plt.show()
```

```python
df = df.drop("SNo", axis=1)
df.plot()
plt.show()
```

```python
turkiye = df[df["Country/Region"]=="Turkey"]
turkiye.columns
```

```python
plt.plot(turkiye.Deaths, turkiye.Recovered, color="red", label="Türkiye'de Ölen-İyileşen Hasta Sayıları")
plt.xlabel("Ölüm Sayısı")
plt.ylabel("İyileşen Hasta Sayısı")
plt.title("Türkiye'deki Coronavirüs Analizi")
plt.legend()
plt.show()
```

```python
italya = df[df["Country/Region"]=="Italy"]
plt.plot(italya.Deaths, italya.Recovered, color="blue", label="İtalya'da Ölen-İyileşen Hasta Sayıları")
plt.xlabel("Ölüm Sayısı")
plt.ylabel("İyileşen Hasta Sayısı")
plt.title("İtalya'daki Coronavirüs Analizi")
plt.legend()
plt.show()
```

```python
turkiye = df[df["Country/Region"]=="Turkey"]
italya = df[df["Country/Region"]=="Italy"]
```

```python
df.plot(subplots=True)
plt.subplot(2,1,1)
plt.plot(turkiye.Deaths, turkiye.Recovered, color="red", label="Türkiye'de Ölen-İyileşen Hasta Sayıları")
plt.xlabel("Ölüm Sayısı")
plt.ylabel("İyileşen Hasta Sayısı")

plt.subplot(2,1,2)
plt.plot(italya.Deaths, italya.Recovered, color="blue", label="İtalya'da Ölen-İyileşen Hasta Sayıları")
plt.xlabel("Ölüm Sayısı")
plt.ylabel("İyileşen Hasta Sayısı")

plt.show()
```

```python
plt.scatter(turkiye.Deaths, turkiye.Recovered, color="red", label="Türkiye'de Ölen-İyileşen Hasta Sayıları")
plt.scatter(italya.Deaths, italya.Recovered, color="blue", label="İtalya'da Ölen-İyileşen Hasta Sayıları")

plt.xlabel("Ölüm Sayısı")
plt.ylabel("İyileşen Hasta Sayısı")
plt.title("Karşılaştırmalı Coronavirüs Analizi")
plt.legend()
plt.show()
```
